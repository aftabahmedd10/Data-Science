#### Q1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Seaborn is a Python data visualization library based on Matplotlib that provides a high-level interface for drawing attractive statistical graphics. Here are five types of plots that can be created using Seaborn, along with their typical use cases:

##### Scatter Plot (sns.scatterplot):

* Use: Display the relationship between two continuous variables.
* Example: Visualizing the correlation between the heights and weights of individuals.

##### Bar Plot (sns.barplot):

* Use: Show the relationship between a categorical variable and a continuous variable.
* Example: Comparing the average sales across different product categories.

##### Histogram (sns.histplot):

* Use: Visualize the distribution of a single variable.
* Example: Analyzing the distribution of exam scores in a class.

##### Box Plot (sns.boxplot):

* Use: Summarize the distribution of a dataset and identify outliers.
* Example: Comparing the distribution of income across different job roles.

##### Heatmap (sns.heatmap):

* Use: Display the correlation matrix or visualize the intensity of a two-dimensional dataset.
* Example: Showing the correlation between various features in a dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Sample data for demonstration
np.random.seed(42)
data = pd.DataFrame({
    'Height': np.random.normal(170, 10, 100),
    'Weight': np.random.normal(70, 5, 100),
    'Category': np.random.choice(['A', 'B', 'C'], 100),
    'Scores': np.random.normal(50, 10, 100)
})
data
```

```python
# Scatter Plot
sns.scatterplot(x='Height', y='Weight', data=data)
plt.title('Scatter Plot')
plt.show()
```

```python
# Bar Plot
sns.barplot(x='Category', y='Scores', data=data)
plt.title('Bar Plot')
plt.show()
```

```python
# Histogram
sns.histplot(data['Scores'], bins=20, kde=True)
plt.title('Histogram')
plt.show()
```

```python
# Box Plot
sns.boxplot(x='Category', y='Height', data=data)
plt.title('Box Plot')
plt.show()
```

```python
# Heatmap
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Heatmap')
plt.show()
```

#### Q2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset
fmri_data = sns.load_dataset("fmri")

# Plot a line plot
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri_data)

# Set plot labels and title
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.title("Line Plot for fmri Dataset")

# Show the plot
plt.show()
```

#### Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.

Note: pclass, age, and fare are columns in the titanic dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset
titanic_data = sns.load_dataset("titanic")

# Plot box plots for 'age' and 'fare' with respect to 'pclass'
plt.figure(figsize=(12, 6))

# Box plot for 'age'
plt.subplot(1, 2, 1)
sns.boxplot(x='pclass', y='age', data=titanic_data)
plt.title('Box Plot for Age by Pclass')

# Box plot for 'fare'
plt.subplot(1, 2, 2)
sns.boxplot(x='pclass', y='fare', data=titanic_data)
plt.title('Box Plot for Fare by Pclass')

# Show the plots
plt.show()
```

#### Q4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset
diamonds_data = sns.load_dataset("diamonds")

# Plot a histogram for 'price' with 'cut' as hue
sns.histplot(data=diamonds_data, x='price', hue='cut', bins=30, kde=True)

# Set plot title and labels
plt.title('Histogram of Diamond Prices by Cut')
plt.xlabel('Price')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

#### Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "iris" dataset
iris_data = sns.load_dataset("iris")

# Plot a pair plot with 'species' as hue
sns.pairplot(iris_data, hue='species')

# Set plot title
plt.suptitle('Pair Plot of Iris Dataset by Species', y=1.02)

# Show the plot
plt.show()
```

#### Q6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset
flights_data = sns.load_dataset("flights")

flights_data
```

```python
# Pivot the data to create a matrix suitable for a heatmap
flights_pivot = flights_data.pivot_table(index='month', columns='year', values='passengers')
flights_pivot
```

```python
# Plot a heatmap
sns.heatmap(flights_pivot, cmap='Blues', annot=True, fmt='d', linewidths=.5)

# Set plot title and labels
plt.title('Passenger Count Heatmap (Flights Dataset)')
plt.xlabel('Year')
plt.ylabel('Month')

# Show the plot
plt.show()
```
