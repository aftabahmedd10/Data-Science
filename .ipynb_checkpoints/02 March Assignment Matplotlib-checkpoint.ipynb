#### Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib. 

##### Matplotlib:
Matplotlib is a 2D plotting library for Python that produces high-quality static, animated, and interactive visualizations in Python. It is one of the most widely used plotting libraries and provides a flexible platform for creating various types of plots and charts.

##### Why is it used:
Matplotlib is used for data visualization, making it easier to understand trends, patterns, and relationships in the data. It is particularly helpful in scientific computing, engineering, data analysis, and machine learning.

##### Five plots that can be plotted using Pyplot module:
The pyplot module in Matplotlib provides a simple interface to create a variety of plots. Here are five types of plots that can be created using pyplot:

```python
#Line Plot:
#Display data points connected by straight line segments.
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.plot(x, y)
plt.show()
```

```python
#Scatter Plot:
#Display individual data points on a two-dimensional plane.
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.scatter(x, y)
plt.show()
```

```python
#Bar Plot:
#Display rectangular bars to represent data.
import matplotlib.pyplot as plt
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]
plt.bar(categories, values)
plt.show()
```

```python
#Histogram:
#Display the distribution of a dataset.
import matplotlib.pyplot as plt
data = [1, 2, 2, 3, 3, 3, 4, 4, 5]
plt.hist(data, bins=5)
plt.show()
```

```python
#Pie Chart:
#Display data as slices of a circular pie.
import matplotlib.pyplot as plt
labels = ['Category A', 'Category B', 'Category C']
sizes = [30, 40, 30]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.show()
```

#### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
Note: Also add title, xlabel, and ylabel to the plot.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

By setting the seed to 3 (np.random.seed(3)), you ensure that every time you run this code, you get the same set of random numbers. This is useful when you want to share code with others or reproduce the same results in different runs. The choice of the seed value (3 in this case) is arbitrary.

```python
# Plot scatter plot
plt.scatter(x, y)

# Add title and labels
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show the plot
plt.show()
```

#### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:

The subplot() function in Matplotlib is used to create multiple plots in the same figure. It allows you to arrange multiple plots in a grid, specifying the number of rows and columns.

Here's how you can use subplot() to draw four line plots:

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for four line plots
x = np.array([0, 1, 2, 3, 4, 5])

y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create subplots
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('Line Plot 1')

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('Line Plot 2')

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('Line Plot 3')

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('Line Plot 4')

# Adjust layout for better appearance
plt.tight_layout()

# Show the plots
plt.show()
```

#### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
# Bar plot
plt.figure(figsize=(8, 6))
plt.bar(company, profit, color='skyblue')
plt.title('Bar Plot - Company Profits')
plt.xlabel('Company')
plt.ylabel('Profit (in billions)')
plt.show()
```

```python
# Horizontal bar plot
plt.figure(figsize=(8, 6))
plt.barh(company, profit, color='lightcoral')
plt.title('Horizontal Bar Plot - Company Profits')
plt.xlabel('Profit (in billions)')
plt.ylabel('Company')
plt.show()
```

#### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It displays the minimum, first quartile (Q1), median, third quartile (Q3), and maximum of a dataset. Box plots are useful for visualizing the spread and skewness of the data.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create a list of data to be plotted
data = [box1, box2]

# Create a box plot
plt.figure(figsize=(8, 6))
plt.boxplot(data, labels=['Box 1', 'Box 2'])
plt.title('Box Plot - Comparison of Two Datasets')
plt.xlabel('Box')
plt.ylabel('Values')
plt.show()
```
